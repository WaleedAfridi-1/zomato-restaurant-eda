## Online Delivery vs. Offline Restaurants

This bar chart compares the number of restaurants that offer **online delivery** versus those that do not.

**Insights:**
- The dataset shows a clear difference between online-enabled restaurants and those without online delivery.
- Restaurants with online delivery represent a notable portion of the market, indicating strong adoption of digital ordering platforms.
- However, a significant share of restaurants still operate offline, suggesting untapped potential for expanding online services.

Exact counts are displayed above each bar for clarity.  
The grid lines along the y-axis make it easier to compare the two categories visually.


```python

delivery=df.value_counts('Has Online delivery').reset_index(name='count')

plt.figure(figsize=(10,8))
sns.barplot(data=delivery,x='Has Online delivery',y='count',
            hue='Has Online delivery',zorder=3)

plt.title("Online Delivery vs. Offline Restaurants", fontsize=16,fontweight='bold')
plt.xlabel("Has Online Delivery?",fontsize=13)
plt.ylabel("Number of Restaurants",fontsize=13)

for i,v in enumerate(delivery['count']):
    plt.text(i, v+50, str(v), ha='center', fontsize=10)
    
plt.grid(True, linestyle='--', alpha=0.5, axis='y')
plt.savefig('barplot_Has_online_deliver.png',dpi=300,bbox_inches='tight')
plt.show()
```

## Top 10 Cuisines Served

This horizontal bar chart highlights the **top 10 most common cuisines** offered by restaurants in the dataset.

**Insights:**
- The most frequently served cuisine appears at the top, showing its dominance in the restaurant market.

# ðŸ’° Average Cost for Two in Top 10 Cities
- This bar chart visualizes the average cost for two people in the top 10 cities with the highest number of restaurants in the dataset.

`ðŸ“Œ Data Processing Steps:`
`Top Cities Identification:`

- First, the top 10 cities were selected based on restaurant count.

`Data Filtering:`

- Dataset was filtered to only include restaurants from these top cities.

`Average Calculation:`

- each city, the mean average cost for two was computed.

`Sorting:`

- Cities were sorted in descending order of average cost to highlight the most expensive markets first.

`ðŸ“Š Plot Details:`
Bar colors: Each city has its own unique color from the "deep" palette.

- X-axis: City names.

- Y-axis: Average cost for two (in local currency).

- Annotations: Exact cost values are displayed above each bar for clarity.

- Gridlines: Horizontal dashed grid lines help compare values across cities.

`ðŸ” Insights:`
- Quickly reveals which cities have the highest dining costs on average.

- Allows for a cost comparison between different urban markets.

- Budget-friendly cities can be spotted instantly, as their bars are shorter.

```python
top_city=df.value_counts("City").nlargest(10).index
feltering=df[df['City'].isin(top_city)]
avg_cost=feltering.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12,8))
sns.barplot(x='City',y='Average Cost for two',data=avg_cost,
           hue='City', palette='deep',zorder=3)

plt.title("Average Cost for Two in Top 10 Cities", fontsize=16, fontweight='bold')
plt.xlabel("City", fontsize=13)
plt.ylabel("Average Cost for Two", fontsize=13)

for i,v in enumerate(avg_cost['Average Cost for two']):
    plt.text(i, v+10, f'{v:.0f}', ha='center', fontsize=10)

plt.grid(linestyle='--', alpha=0.5, axis='y')
plt.tight_layout()
plt.xticks(rotation=45)
plt.savefig('Average Cost for Two in Top 10 Cities.png',dpi=300,bbox_inches='tight')
plt.show()
```

# ðŸ™ City-wise Price Range vs Average Cost for Two
- This heatmap compares the average cost for two people across different price ranges in the top 10 cities with the most restaurants in the dataset.

- `ðŸ“Œ Data Processing Steps:`
- `Top Cities Selection:`

- We first identified the top 10 cities based on the highest restaurant counts.

`Grouping:`

- Data was grouped by City and Price range to calculate the mean average cost for two.

`Pivot Table Creation:`

- The grouped data was reshaped so that rows represent cities, columns represent price ranges, and cell values represent the average cost.

`NaN Handling:`

- Missing values were replaced with 0 to avoid blank spots in the heatmap.

`ðŸ“Š Plot Details:`
- Color coding: Darker shades represent higher average costs, lighter shades indicate lower costs.

- Annotations: Each cell shows the exact average cost for two (rounded to nearest integer).

- X-axis: Price ranges from 1 (cheap) to 4 (expensive).

- Y-axis: Top 10 cities.

`ðŸ” Insights:`
- You can easily see which price range is more dominant in each city.

- Some cities may have low average cost even in higher price ranges, indicating affordability.

- Others may show very high costs in premium ranges, highlighting expensive dining markets.

```python
top_city=df.value_counts('City').nlargest(10).index
f_data=df[df['City'].isin(top_city)]

groupdata=f_data.groupby(['City','Price range'])['Average Cost for two'].mean().reset_index()
pivot_table=groupdata.pivot(index='City',columns='Price range',values='Average Cost for two')
pivot_table.fillna(0,inplace=True)

plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, fmt='.0f', annot=True,
            cmap='YlGnBu',cbar_kws={'label': 'Average Cost for Two'}
           )

plt.title("City-wise Price Range vs Average Cost for Two", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("Price Range (1 = Cheap, 4 = Expensive)", fontsize=12)
plt.ylabel("City", fontsize=12)

plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('heatmap_City-wise Price Range vs Average Cost for Two.png',dpi=300,bbox_inches='tight')
plt.show()
```

# ðŸ“Š Correlation Heatmap â€” Numerical Columns
- This heatmap visualizes the correlation between all numerical features in the dataset.

- `Color scale:`

- Dark red (close to +1) â†’ Strong positive correlation (both variables increase together).

- Dark blue (close to -1) â†’ Strong negative correlation (one increases while the other decreases).

- Near white (around 0) â†’ Weak or no correlation.

- `ðŸ” Insights:`

# Pair Plot: Cost, Rating & Location (Hue = Price Range)
- `This pair plot visualizes the relationship between five key features of the Zomato restaurant dataset:`

- Average Cost for Two â€“ The typical cost for two people dining.

- Price Range â€“ A categorical rating from 1 (cheapest) to 4 (most expensive).

- Aggregate Rating â€“ Average customer rating on Zomato.

- Longitude & Latitude â€“ Restaurant location coordinates.

- `Key points about the plot:`

- The diagonal plots show the distribution (KDE curves) of each variable individually.

- The off-diagonal scatter plots compare variables in pairs, with each point colored by Price Range.

- `This helps in spotting:`

- Outliers â€” e.g., a Price Range 1 restaurant with very high costs.

- Patterns â€” e.g., higher price range restaurants tending towards higher ratings.

- Geographical clusters â€” locations where certain price ranges dominate.

- `Why this matters:`
- This visualization combines multiple relationships in a single figure, making it easier to identify trends, correlations, and anomalies between cost, pricing category, ratings, and location.

```python

sns.pairplot(df[['Average Cost for two','Price range', 'Aggregate rating', 'Longitude', 'Latitude']], diag_kind='kde',
             hue='Price range',palette='pastel',height=2.5)


plt.suptitle("Pair Plot: Cost, Rating & Location (Hue = Price Range)",
             fontsize=16, fontweight='bold', y=1.02)

plt.savefig("pairplot_price_range.png", dpi=300, bbox_inches='tight')
plt.show()
```
